```python
import itertools

# Define the list of elements
elements = ['A', 'B', 'C', 'D', 'E']

def display_combinations(elements):
    """Generate and display combinations of varying lengths."""
    print("\nGenerating Combinations:")
    for length in range(1, len(elements) + 1):
        combs = list(itertools.combinations(elements, length))
        print(f"Combinations of length {length}: {combs}")

def display_permutations(elements):
    """Generate and display permutations of varying lengths."""
    print("\nGenerating Permutations:")
    for length in range(1, len(elements) + 1):
        perms = list(itertools.permutations(elements, length))
        print(f"Permutations of length {length}: {perms}")

# Execute functions to display combinations and permutations
display_combinations(elements)
display_permutations(elements)
```

```python
import itertools

# List of elements
items = ['A', 'B', 'C', 'D', 'E']

def generate_combinations(elements):
    """
    Generate all possible combinations of elements for different lengths.
    """
    results = {}
    for length in range(1, len(elements) + 1):
        # Create a dictionary entry for each length
        results[length] = list(itertools.combinations(elements, length))
    return results

def generate_permutations(elements):
    """
    Generate all possible permutations of elements for different lengths.
    """
    results = {}
    for length in range(1, len(elements) + 1):
        # Create a dictionary entry for each length
        results[length] = list(itertools.permutations(elements, length))
    return results

def display_results(title, results):
    """
    Print results with a given title.
    """
    print(f"\n{title}:")
    for length, items in results.items():
        print(f"\nLength {length}:")
        for item in items:
            print(item)

# Compute combinations and permutations
combination_results = generate_combinations(items)
permutation_results = generate_permutations(items)

# Display the results
display_results("Combinations", combination_results)
display_results("Permutations", permutation_results)
```
